```python
#Dependencies and read in data file
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb


city_df = pd.read_csv("city_data.csv")
ride_df = pd.read_csv("ride_data.csv")
```

** Both tables above have 125 unique cities.  As long as there are 125 cities at the end - every city is represented.**


**Next section deals withe the 2 instances of Port James - the totals are summed.  There is probably a much easier way to do this...**

```python
citycount = city_df["city"].value_counts()
citycount
```

```python
city_df_sum = city_df.groupby(["city"]).sum()
city_df_sum.reset_index(inplace=True)
city_df_sum
```

```python
city_df_sum['city'].replace({'Port James': 'Port James L'},inplace=True)
city_df_sum = pd.DataFrame(city_df_sum)
city_df_sum
```

```python
city_df_sum.rename(columns={"driver_count":"driver_count_L"},inplace=True)
city_df_sum
```

```python
city_df = pd.merge(city_df_sum, city_df, on="city", how="left")
city_df
```

```python
city_df = city_df.drop(["driver_count"],axis=1)
city_df
```

**Above is the final city_df table which contains a list of all 125 cities - their city type - and the number of drivers in 
the city.  **

** Now begin working on the ride_df table.  This table gives number of rides taken for each city along with total rides. **

```python
ride_sum = ride_df.groupby("city").sum()
ride_count = ride_df.groupby("city").count()
ride_sum
```

```python
ride_sum.reset_index(inplace=True)
ride_sum
```

```python
ride_sum = ride_sum.drop(["ride_id"],axis=1)
ride_sum
```

** Ride Sum table above has sum of fares by city (125 unique cities) **

```python
ride_count
```

```python
ride_count = ride_count.drop(["fare","ride_id"],axis=1)
ride_count
```

```python
ride_count.rename(columns={"date":"total rides"},inplace=True)
ride_count
```

```python
ride_count.reset_index(inplace=True)
ride_count
```

** Ride count table has number of rides by city (125 cities). **

** Now to put all 3 table together (city_df, ride count and ride sum) **

```python
city_df = pd.merge(city_df, ride_count, on="city", how="left")
city_df
```

```python
city_df = pd.merge(ride_sum, city_df, on="city", how="left")
city_df
```

```python
city_df["average fare"] = city_df["fare"]/city_df["total rides"]
city_df
```

```python
Suburban = city_df.loc[city_df['type'] == "Suburban"]
Suburban
```

```python
Urban = city_df.loc[city_df['type'] == "Urban"]
Urban
```

```python
Rural = city_df.loc[city_df['type'] == "Rural"]
Rural
```

```python
Urban_plot = plt.scatter(Urban["total rides"], Urban["average fare"], marker="o", facecolors="lightcoral", edgecolors="black",
            s=Urban["driver_count_L"]*5, alpha=0.75,label = "Urban")
Suburban_plot = plt.scatter(Rural["total rides"], Rural["average fare"], marker="o", facecolors="lightblue", edgecolors="black",
            s=Rural["driver_count_L"]*5, alpha=0.75,label = "Rural")
Rural_plot = plt.scatter(Suburban["total rides"], Suburban["average fare"], marker="o", facecolors="gold", edgecolors="black",
            s=Suburban["driver_count_L"]*5, alpha=0.75,label = "Suburban")
plt.legend(handles=[Urban_plot, Rural_plot, Suburban_plot], loc="best")
plt.xlabel("Total Rides")
plt.ylabel("Average Fare")
plt.show()
```

**In the Scatter Plot above - size of bubbles is correlated with driver count. **

```python
fares_by_type = city_df.groupby("type").sum()
fares_by_type.reset_index(inplace=True)
fares_by_type
```

```python
fares_by_type["fares per driver"] = fares_by_type["fare"]/fares_by_type["driver_count_L"]
fares_by_type["fares per trip"] = fares_by_type["fare"]/fares_by_type["total rides"]
fares_by_type
```

```python
#colors = ["yellowgreen", "red", "lightcoral"]
colors = ["Gold", "lightblue", "lightcoral"]
plt.title("Total Fare by City Type")
plt.pie(fares_by_type["fare"], labels=fares_by_type["type"], colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.show()
```

```python
plt.title("Total Rides by City Type")
plt.pie(fares_by_type["total rides"], labels=fares_by_type["type"], colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.show()
```

```python
plt.title("Driver Count by City Type")
plt.pie(fares_by_type["driver_count_L"], labels=fares_by_type["type"], colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.show()
```

** Observation 1:  Majority of Pyber business is in Urban. **

** Observation 2:  Highest average fares appear to be in Rural. **
    
** Observation 3:  Urban cities have the highest driver counts.  **
