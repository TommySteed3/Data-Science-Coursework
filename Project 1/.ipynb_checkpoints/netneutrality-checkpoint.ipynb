```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
import codecs, json, time, re, tweepy
```

```python
with codecs.open('../Scrape_Twitter_for_Tweets/twitterscraper/netneutrality.json', 'r', 'utf-8') as f:
    tweets = json.load(f, encoding='utf-8')
```

```python
df = pd.DataFrame(tweets)
df
```

```python
# Parsing timestamp into Date & Time columns - T is the Time component in ISO 8601 
for time in df["timestamp"]:
    for i in np.arange(len(df["timestamp"])):
        df["Date(UTC)"] = df["timestamp"][i].split("T")[0]
        df["Time(UTC)"] = df["timestamp"][i].split("T")[1]
        
df
```
