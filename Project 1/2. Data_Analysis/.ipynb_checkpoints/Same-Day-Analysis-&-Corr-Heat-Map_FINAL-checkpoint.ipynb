# PART II: MARKET ANALYSIS

```python
from pandas_datareader import data
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# We would like all available data from MM/DD/YYYY until MM/DD/YYYY.
start_date = target_terms_df["Date"].min().strftime('%Y-%m-%d')
end_date = target_terms_df["Date"].max().strftime('%Y-%m-%d')
print("Start date is: "+start_date)
print("End date is: "+end_date)
```

```python
# Define the instruments to download. We would like to see Apple, Microsoft and the S&P500 index.
tickers = ['^DJI', '^IXIC', '^NYA']#, 'SNP', 'XOM']

# Define which online source one should use
data_source = 'yahoo'

# User pandas_reader.data.DataReader to load the desired data. As simple as that.
panel_data = data.DataReader(tickers, data_source, start_date, end_date)
```

```python
# Getting just the adjusted closing prices. This will return a Pandas DataFrame
# The index in this DataFrame is the major index of the panel_data.
adj_close = panel_data.ix['Adj Close']

# Getting all weekdays between MM/DD/YYYY and MM/DD/YYYY
all_weekdays = pd.date_range(start=start_date, end=end_date, freq='B')

# How do we align the existing prices in adj_close with our new set of dates?
# All we need to do is reindex close using all_weekdays as the new index
adj_close = adj_close.reindex(all_weekdays)

# Reindexing will insert missing values (NaN) for the dates that were not present
# in the original set. To cope with this, we can fill the missing by replacing them
# with the latest available price for each instrument.
adj_close = adj_close.fillna(method='ffill')

adj_close.head()
```

```python
adj_close["DJI Returns"] = ""
adj_close["IXIC Returns"] = ""
adj_close["NYA Returns"] = ""
adj_close.head()
```

```python
adj_close.index.name = "Date"
adj_close = adj_close.reset_index()
adj_close.head()
```

```python
# Set initial return value as this is the start
adj_close.set_value(0, "DJI Returns", 0)
adj_close.set_value(0, "IXIC Returns", 0)
adj_close.set_value(0, "NYA Returns", 0)
i = 0

# Scaling Factor for comparison with Twitter
m = 50 

# Iterate through each row and calculate the daily returns
for index, row in adj_close.iterrows():
    if index==0:
        print("1st row skipped")
    else:
        try:
            # Calculate the daily returns
            dji_returns = ((row["^DJI"]/adj_close["^DJI"][i])-1)*m
            ixic_returns = ((row["^IXIC"]/adj_close["^IXIC"][i])-1)*m
            nya_returns = ((row["^NYA"]/adj_close["^NYA"][i])-1)*m            

            # Set return values to appropriate index
            adj_close.set_value(index, "DJI Returns", dji_returns)
            adj_close.set_value(index, "IXIC Returns", ixic_returns)
            adj_close.set_value(index, "NYA Returns", nya_returns)
            i += 1
            print(".", end="")
        except:
            print("This doesn't work. Please try again")

adj_close_returns = adj_close.set_index("Date")
adj_close_returns.head()
target_terms_df1 = target_terms_df.set_index("Date")
```

```python
result = pd.merge(target_terms_df1, adj_close_returns, how='outer', on=None, left_on=None, right_on=None,
         left_index=True, right_index=True)
result = result.fillna(method='ffill').rename(columns={"Compound_stockmarket":"'stockmarket' cmpnd","target dow":"'Dow' cmpnd","target nasdaq":"'NASDAQ' cmpnd","target nyse":"'NYSE' cmpnd","DJI Returns": "Dow Returns","IXIC Returns": "NASDAQ Returns","NYA Returns": "NYSE Returns"})
result1 = result.drop(["^DJI","^IXIC","^NYA"], axis=1)
```
