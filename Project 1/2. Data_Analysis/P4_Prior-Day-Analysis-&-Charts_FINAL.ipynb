```python
from pandas_datareader import data
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from datetime import datetime
```

```python
# TABLE OF STOCK RETURNS & SENTIMENT VALUES
combined_table = pd.read_csv("../1 Data_Collection_Processing/combined_table.csv")
```

# CORRELATION PLOT

```python
xvalues = list(range(1, 31))
width = 60
height = 15

# CREATE TABLE FOR PLOTTING
fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize=(width,height))

sns.set(font_scale=5)
sns.set_style("darkgrid")

# NEUTRAL
sns.regplot(combined_table["prior paul"], 
            combined_table["Daily XAX"],
            fit_reg=True, ci=95, n_boot=1000,
            order=1, ax = ax[0],
            scatter_kws={"s": 1000, 'color':'darkblue'},
            line_kws={"color":"darkblue","alpha":0.75,"lw":9})
ax[0].grid(b=True, which='major', color='k', linestyle='-.', lw = 1)
ax[0].set_ylim(bottom= -0.015, top = 0.010)    
ax[0].set_xlabel("Paul Krugman Sentiment")
ax[0].set_ylabel("Daily XAX")
ax[0].set_title("No Correlation")

ticklines = ax[0].get_xticklines() + ax[0].get_yticklines()
# gridlines = ax.get_xgridlines() + ax.get_ygridlines()
# ticklabels = ax.get_xticklabels() + ax.get_yticklabels()

# for line in ticklines:
#     line.set_linewidth(3)

# for line in gridlines:
#     line.set_linestyle('-.')
#     line.set_color('k')

# for label in ticklabels:
#     label.set_color('k')
#     label.set_fontsize('large')

# for tick in ax[0].get_xticklabels():
#     tick.set_rotation(0)
#     tick.set_fontsize(16)


# NEGATIVE
sns.regplot(combined_table["prior paul"], 
            combined_table["Daily RUT"], 
            fit_reg=True, ci=95, n_boot=1000,
            order=1, ax = ax[1],
            scatter_kws={"s": 1000, 'color':'goldenrod'},
            line_kws={'color':'goldenrod',"alpha":0.75,"lw":9})

ax[1].grid(b=True, which='major', color='k', linestyle='-.')
ax[1].set_ylim(bottom= -0.015, top = 0.010)    
ax[1].set_xlabel("Paul Krugman Sentiment")
ax[1].set_ylabel("Daily RUT")
ax[1].set_title("Negative Correlation")

# POSITIVE
sns.regplot(combined_table["prior justin"], 
            combined_table["Daily XAX"],
            fit_reg=True, ci=95, n_boot=1000,
            order=1, ax = ax[2],
            scatter_kws={"s": 1000, 'color':'darkred'},
            line_kws={'color':'darkred',"alpha":0.75,"lw":9})

ax[2].grid(b=True, which='major', color='k', linestyle='-.')
ax[2].set_ylim(bottom= -0.015, top = 0.010)        
ax[2].set_xlabel("Justin Sentiment")
ax[2].set_ylabel("Daily XAX")
ax[2].set_title("Postive Correlation")
plt.savefig('3-type-corr3.png', facecolor ='0.95', 
            bbox_inches="tight",
            transparent=True, frameon=True
           )
plt.show()
```

# BENCH MARK TEST OF PREDICTABILITY

```python
import statsmodels.api as sm # import statsmodels 

X = combined_table["prior justin"] ## X usually means our input variables (or independent variables)
y = combined_table["Daily XAX"] ## Y usually means our output/dependent variable
X = sm.add_constant(X) ## let's add an intercept (beta_0) to our model

# Note the difference in argument order
model = sm.OLS(y, X).fit() ## sm.OLS(output, input)
predictions = model.predict(X)

combined_table["Predicted"] = predictions

# Print out the statistics
model.summary()
```

# STOCK & TWEET SENTIMENT PLOT: STOCK ANALYST JUSTIN WOLFER

```python
import numpy as np

width = 10
height = 10
sns.set(font_scale=1)
fig = plt.figure(figsize=(width,height))

ax = fig.add_subplot(1,1,1)
ax.plot(pd.to_datetime(combined_table["Date"]), 
        combined_table["Daily XAX"]*100,
        marker=".", linewidth=3, ms = 10,ls = "--",
        label = "NYSE", color = "darkgreen"
        )
ax.plot(combined_table["Date"], 
        combined_table["prior justin"],
        marker="*", linewidth=3, ms = 10,
        label = "Justin Wolfers", color = "orangered"
        )
ax.grid(True)

ticklines = ax.get_xticklines() + ax.get_yticklines()
gridlines = ax.get_xgridlines() + ax.get_ygridlines()
ticklabels = ax.get_xticklabels() + ax.get_yticklabels()

for line in ticklines:
    line.set_linewidth(3)

for line in gridlines:
    line.set_linestyle('-.')
    line.set_color('k')

for label in ticklabels:
    label.set_color('k')
    label.set_fontsize('large')

for tick in ax.get_xticklabels():
    tick.set_rotation(0)

ax.set_xlabel("Market Day", fontsize = "large")
ax.set_ylabel("Sentiment/Performance", fontsize = "large")
ax.set_title("Justin Wolfer's \n 1-Day Before Sentiment & NYSE Returns", fontsize = "xx-large")
ax.legend()

plt.savefig('wolf_sentiment_NYSE.png', 
            bbox_inches='tight', 
            dpi=200, 
            transparent=True)
plt.show()
```

# PREDICTABILITY PLOT

```python
width = 10
height = 10
sns.set(font_scale=1)
fig = plt.figure(figsize=(width,height))

ax = fig.add_subplot(1,1,1)
ax.plot(pd.to_datetime(combined_table["Date"]), 
        combined_table["Daily XAX"]*100,
        marker=".", linewidth=3.5, ms = 10,ls = "--",
        label = "NYSE", color = "darkgreen"
        )
ax.plot(combined_table["Date"], 
        combined_table["Predicted"]*100,
        marker="*", linewidth=3.5, ms = 10,
        label = "Predicted Value", color = "orangered"
        )
ax.grid(True)

ticklines = ax.get_xticklines() + ax.get_yticklines()
gridlines = ax.get_xgridlines() + ax.get_ygridlines()
ticklabels = ax.get_xticklabels() + ax.get_yticklabels()

for line in ticklines:
    line.set_linewidth(3)

for line in gridlines:
    line.set_linestyle('-.')
    line.set_color('k')

for label in ticklabels:
    label.set_color('k')
    label.set_fontsize('x-large')

for tick in ax.get_xticklabels():
    tick.set_rotation(0)

ax.set_xlabel("Market Day", fontsize = "x-large")
ax.set_ylabel("Sentiment/Performance", fontsize = "xx-large")

ax.set_title("Justin Wolfer's \n Predictive Power & NYSE Returns", fontsize = "xx-large")
ax.legend()

plt.savefig('wolfs_predictive_modeling.png', 
            bbox_inches='tight', 
            dpi=200, 
            transparent=True)
plt.show()
```
